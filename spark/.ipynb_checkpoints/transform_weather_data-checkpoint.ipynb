```python
from datetime import date, datetime
from pyspark.sql.types import *
from pyspark.sql.functions import *
import datetime as dt
import pandas as pd
import json
from pyspark.context import SparkContext
from pyspark.sql.session import SparkSession
# sc = SparkContext('local')
# spark = SparkSession(sc)
```

```python
def cal_std_day(befor_day):   
    x = dt.datetime.now() - dt.timedelta(befor_day)
    year = x.year
    month = x.month if x.month >= 10 else '0'+ str(x.month)
    day = x.day if x.day >= 10 else '0'+ str(x.day)  
    return str(year)+str(month)+ str(day)
```

## Transform

```python
path = '/final_data/weather/weather_20220417.json'
tmp = spark.read.json(path, encoding='utf-8')
tmp.show()
```

```python
tmp2 = tmp.select('response').first()
df = spark.createDataFrame(tmp2)
tmp3 = df.select('body').first()
tmp4 = spark.createDataFrame(tmp3)
tmp5 = tmp4.select('items').first()
df2 = spark.createDataFrame(tmp5).first()
fin_df = spark.createDataFrame(df2['item'])
fin_df.show()
```

```python
weather = fin_df.select(substring(col('tm'),1,10).alias('날짜')
                           ,substring(col('tm'),12,17).alias('시간대')
                          ,col('ta').alias('기온(°C)')
                          ,col('rn').alias('강수량(mm)')
                          ,col('ws').alias('풍속(m/s)')
                          ,col('wd').alias('풍향(16방위)')
                          ,col('hm').alias('습도(%)')
                          ,col('pv').alias('증기압(hPa)')
                          ,col('td').alias('이슬점온도(°C)')
                          ,col('pa').alias('현지기압(hPa)')
                          ,col('ps').alias('해면기압(hPa)')
                          ,col('ss').alias('일조(hr)')
                          ,col('icsr').alias('일사(MJ/m2)')
                          ,col('dsnw').alias('적설(cm)')
                          ,col('vs').alias('시정(10m)')
                          ,col('ts').alias('지면온도(°C)')).filter((col('시간대') >= '10:00') & (col('시간대') <= '22:00'))
                        
weather.show()
```
